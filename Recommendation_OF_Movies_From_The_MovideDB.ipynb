```python
df_list = []
```

```python
df = pd.DataFrame()
```

```python
for i in range(1,3):
    response = requests.get('https://api.themoviedb.org/3/movie/581babb8654e44e2e5c92e81c8745eb6/recommendations?api_key=581babb8654e44e2e5c92e81c8745eb6&language=en-US&page={}'.format(i))
    temporary_df = pd.DataFrame(response.json()['results'])
    temporary_df = temporary_df[['id','title','overview','original_language','popularity','release_date','vote_average','vote_count']]
    df_list.append(temporary_df)
```

```python
df = pd.concat(df_list, ignore_index=True)
```
